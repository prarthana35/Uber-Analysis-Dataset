DATA PREPROCESSING

```python
data['PURPOSE'].fillna("Not", inplace = True)  #change nan to not
```

```python
data.head(20)
```

```python
data['START_DATE'] = pd.to_datetime(data['START_DATE'],errors = 'coerce')
```

```python
data['END_DATE'] = pd.to_datetime(data['END_DATE'], errors='coerce')
```

```python
data.info()
```

```python
from datetime import datetime
data['date'] = pd.DatetimeIndex(data['START_DATE']).date
data['time'] = pd.DatetimeIndex(data['START_DATE']).hour
data.head()
```

```python
data['day-night'] = pd.cut(x=data['time'],bins = [0,10,15,19,24],labels = ['Morning','Afternoon','Evening','Night']) #add column
data.head()
```

```python
data.dropna(inplace = True)
data.shape
data.head()
```

DATA VISUALIZATION

```python
#IN WHICH CATEGORY DO PEOPLE BOOK THE MOST UBER RIDES?
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.countplot(data['CATEGORY'])
plt.xticks(rotation = 90)
plt.subplot(1,2,2)
sns.countplot(data['PURPOSE'])
```

```python
#AT WHAT TIME DO PEOPLE BOOK CABS THE MOST FROM UBER?
sns.countplot(data['day-night'])
```

```python
data['MONTH'] = pd.DatetimeIndex(data['START_DATE']).month
month_label = {1.0:'Jan',2.0:'Feb',3.0:'Mar',4.8:'April',
               5.0:'May',6.0:'June',7.0:'July',8.0:'Aug',
               9.0:'Sep',10.0:'Oct',11.0:'Nov',12.0:'Dec'}

data['MONTH']= data.MONTH.map(month_label)
mon = data.MONTH.value_counts(sort=False)

data.head()
```

```python
data = data.dropna()
data = data.drop(columns=['day-bight'])

data.shape
data.head()
```

```python
#IN WHICH MONTH DO PEOPLE BOOK UBER RIDES LESS FREQUENTLY?
df = pd.DataFrame({
    "MONTH":mon.values,
    "VALUE COUNT": data.groupby("MONTH",sort=False)['MILES'].max()
})
p= sns.lineplot(data=df)
p.set(xlabel= "MONTHS",ylabel="VALUE COUNT")
```

```python
#In which days of weeks do people book the uber rides the most?
data['DAY'] = data.START_DATE.dt.weekday
day_label ={
    0: 'Mon',1: 'Tue',2: 'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}

data['DAY'] = data['DAY'].map(day_label)
data.head()
```

```python
day_label = data.DAY.value_counts()
sns.barplot(x=day_label.index, y=day_label)
plt.xlabel('DAY')
plt.ylabel('COUNT')
```
